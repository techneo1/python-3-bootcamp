```python
# STRINGS
# Strings are a built-in data type in Python
# Strings are immutable sequences of characters
# Strings are enclosed in single or double quotes
# Strings can be indexed and sliced
# Strings can be concatenated using the + operator
# Strings can be repeated using the * operator
# Strings can be split into lists using the split() method
# Strings can be formatted using f-strings or the format() method
# Strings can be iterated over using loops

mystring = "abcdefghijk lmnopqrstuvw xyz" # Assigning a string to a variable
print(mystring) # Printing the string
print(len(mystring)) # Finding the length of the string

# Slicing
print(mystring[:5:2]) # Slicing the string from index 3 to 10 with a step of 2
print(mystring[::2]) # Slicing the string from index 0 to the end with a step of 2
print(mystring[::-1]) # Reversing the string

# Indexing
print(mystring[26]) # Extracting character "y"
print(mystring[-2]) # Extracting characters "y" from reverse string

# IMMUTABILITY
name = "John"
#name[0] = "P"   # TypeError! because strings are IMMUTABLE
print(name.upper()); # JOHN
print(name); # John, Original string is unchanged
 
# Concatenation 
print(name[0] + 'ohny') # 'Johny'
print('2'+'3') # '23'

# Repeatation
print('z' * 4) # zzzz
print('2' * 3) # 222

# Splitting
x = "Here is a string"
x.split(" ") # Splitting the string into a list of words
```

```python
# Formatting
print('This is a string {}'.format("INSERTED")) # Using format() method
print('The {} {} {}'.format("quick", "brown", "fox")) # Using format() method
print('The {2} {1} {0}'.format("fox", "brown", "quick")) # Using format() method
print('The {q} {b} {f}'.format(f="fox", b="brown", q="quick")) # Using format() method

#{value:{width}.{precision}f}
print("The result is {r:10.2f}".format(r=1.23456789)) # Using format() method

name = 'Srikanth'
print('My name is {}'.format(name)) # Using format() method
print('My name is {0}'.format(name)) # Using format() method
print('My name is %s' % name) # Using % operator
print(f'My name is {name}') # Using format() method
```

```python
# LISTS
# Lists are ordered, mutable, and allow duplicate elements
# Lists are defined using square brackets []
# Lists can contain elements of different data types
# Lists can be nested (lists within lists)

my_list  = ["apple", 100, 3.14, True, None]
print(my_list)

len(my_list)

# Just like String INDEXING and SLICING can be performed. Only difference is that it is MUTABLE
nested_list = [1,1,[1,2]]
print(nested_list[2][1])   # Accessing 2 from the nested list

# List methods
my_list.append("banana")
print(my_list)

popped_item = my_list.pop()
print(popped_item)
print(my_list)

popped_item = my_list.pop(4)
print(popped_item)
print(my_list)

new_list = ['e', 'i', 'a', 'u', 'o']
none_type = new_list.sort()
print(new_list)
print(type(none_type))

another_list = [1, 2, 3, 4, 5]
another_list.reverse()
print(another_list)

```
